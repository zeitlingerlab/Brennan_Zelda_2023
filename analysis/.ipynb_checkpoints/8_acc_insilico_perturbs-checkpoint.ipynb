# Introduction

The purpose of this .ipynb is to generate motif-pair in-silico perturbations across canonical `ZDTBCH` motifs and their variants. These perturbations will be measured at the level of accessibility and not binding. This will allow us to assess whether motif pairs have preferential distances of cooperativity or directional hierarchy in how they pioneer/increase accessibility. Additionally, we want to investigate the role of Zelda opening chromatin in the presence of low-affinity Zelda motifs. To do this, we will measure marginalization scores of different Zelda motif variants for ChromBPNet.

# Define scope of in silico perturbations

Here, we will define the consensus sequences for motifs of interest that were returned by the relevant ChIP-nexus BPNet/TF-MoDISco model as well as the distances by which they will be injected in a pairwise fashion.

```python
motifs_dict = {
    'Zld': 'CAGGTAG',
    'Zld-lowaff': 'CAGGTA',
    'Zld-lowestaff': 'TAGGTAG',
    'Gaf': 'GAGAGAGAGAGAGAGAG',
    'Cad': 'TTTTATGGCC',
    'Dl': 'GGGAAAACCC',
    'Twi': 'AACACATGTT',
    'Bcd': 'TTAATCC'
}

timepoints = ['1to15','15to2','2to25','25to3']

#Identify other parameters
trials = 512
correct_for_shoulder_synergy = True
measurement_window_around_motif = 400
```

# Inject motif over randomly generated features to assess their profiles

Here, for each motif that we have identified, we will inject that motif into the center of the profile and measure the predicted value over the null values.

First, generate the random sequences.

```python
def generate_random_seq(seqlen, weights = [.25, .25, .25, .25]):
    """
    Purpose: Generate a random DNA sequence of a specified length.
    """
    import random
    seq = random.choices(['A','C','G','T'], weights = weights, k=seqlen)
    return(''.join(seq))

random.seed(10)
random_seqs = [generate_random_seq(seqlen = input_seqlen) for i in range(trials)]
random_seqs_1he = one_hot_encode_sequences(random_seqs)
random_seqs_1he.shape
```

## In silico injections into random sequences

Next, inject each motif into these random sequences, measure the profile, then average over all trials.

```python
%%script false --no-raise-error
profile_df =  pd.DataFrame()
perturbs_df = pd.DataFrame()

for t in timepoints:
    print(t)
    
    #Import model
    tf.keras.backend.clear_session()
    model_path = f'{model_dir}_{t}.h5'
    bias_model_path = f'{model_dir}_{t}.adjusted_bias_model.h5'
    model_chrombpnet, model_bias = load_chrombpnet(model_path = model_path, bias_model_path = bias_model_path)
    
    for motifA,seqA in tqdm(motifs_dict.items()):

        #Perform in silico injections
        preds_dict = predict_injected_seq(model_chrombpnet = model_chrombpnet, model_bias = model_bias,
                                    primary_motif = '',
                                    primary_position = 500,
                                    secondary_motif = seqA,
                                    secondary_positions = [500],
                                    input_seqlen = input_seqlen ,
                                    input_seqs = random_seqs_1he,
                                    trials = trials)
        
        df = pd.DataFrame(preds_dict['null'], columns = ['signal'])
        df['position'] = list(range(df.shape[0]))
        df['position'] = df['position'] - 500
        df['type'] = 'no motif'
        df['motifA'] = motifA
        df['timepoint'] = t
        profile_df = profile_df.append(df)
        
        df = pd.DataFrame(preds_dict['distances'][500]['AB'], columns = ['signal'])
        df['position'] = list(range(df.shape[0]))
        df['position'] = df['position'] - 500
        df['type'] = 'motif'
        df['motifA'] = motifA
        df['timepoint'] = t        
        profile_df = profile_df.append(df)
    
profile_df.to_csv('tsv/perturbs/accessibility/insilico/insilico_inj_profiles_random_seq.tsv.gz', sep = '\t', index = False)
```

## Injection into in vivo "nonpeak" sequences

Now, we want to measure whether these injections perform similarly on in vivo sequences that should show no accessibilty. First, we need to import the sequences from the non-peak regions

```python
test_peaks_seqs, test_peaks_cts, test_nonpeaks_seqs, test_nonpeaks_cts = \
    load_test_data(peaks_file = peak_bed, 
                   nonpeaks_file = nonpeak_bed, 
                   genome_fasta = reference_genome, 
                   cts_bw_file = cts_bw_file,
                   test_chrs = [test_chrom], 
                   inputlen = input_seqlen, outputlen = output_seqlen)
```

Filter the sequences to be low-count regions. For this case, we want there to be less than 1000 reads per window.

```python
regions_under_1000_reads_idx = np.sum(test_nonpeaks_cts, axis = 1) < 1000
nonpeak_seqs = test_nonpeaks_seqs[regions_under_1000_reads_idx]
nonpeak_seqs.shape
```

This gives us 1125 regions to work with. This is sufficient for averaging across.

Then, define a function to inject motif into a genomic sequence.

```python
def predict_motif_inj_into_seqs(model_chrombpnet, model_bias, seqs, motif):
    w_mot_seqs = seqs.copy()
    MID = (model_chrombpnet.inputs[0].shape[1])//2 - len(motif)//2
    w_mot_seqs[:, MID:MID+len(motif)] = one_hot.dna_to_one_hot([motif])
        
    _, preds_wt = predict_chrombpnet(model_chrombpnet, model_bias, seqs, batch_size=128, verbose = False, no_bias = True)
    _, preds_mut = predict_chrombpnet(model_chrombpnet, model_bias, w_mot_seqs, batch_size=128, verbose = False, no_bias = True)
    
    return preds_wt, preds_mut
```

Next, inject each motif into these random sequences, measure the profile, then average over all trials.

```python
%%script false --no-raise-error
profile_df =  pd.DataFrame()
perturbs_df = pd.DataFrame()

for t in timepoints:
    print(t)
    
    #Import model
    tf.keras.backend.clear_session()
    model_path = f'{model_dir}_{t}.h5'
    bias_model_path = f'{model_dir}_{t}.adjusted_bias_model.h5'
    model_chrombpnet, model_bias = load_chrombpnet(model_path = model_path, bias_model_path = bias_model_path)
    
    for motifA,seqA in tqdm(motifs_dict.items()):
        preds_wt, preds_mut = predict_motif_inj_into_seqs(model_chrombpnet, model_bias, nonpeak_seqs, seqA)
        preds_wt = np.mean(preds_wt, axis = 0)
        preds_mut = np.mean(preds_mut, axis = 0)
        
        wt_df = pd.DataFrame(preds_wt, columns = ['signal'])
        wt_df['position'] = list(range(wt_df.shape[0]))
        wt_df['position'] = wt_df['position'] - 500 
        wt_df['type'] = 'WT'
        mut_df = pd.DataFrame(preds_mut, columns = ['signal'])
        mut_df['position'] = list(range(mut_df.shape[0]))
        mut_df['position'] = mut_df['position'] - 500 
        mut_df['type'] = 'mut'        
        
        df = pd.concat([wt_df, mut_df], axis = 0)
        df['motifA'] = motifA
        df['timepoint'] = t
            
        profile_df = profile_df.append(df)
    
profile_df.to_csv('tsv/perturbs/accessibility/insilico/insilico_inj_profiles_nonpeak_seq.tsv.gz', 
                   sep = '\t', index = False)
```

Above, we generated many .pkl files that contain the information stored in these perturbations. However, we also want to summarize the profiles into general count values for each paired situation. In order to do this, we will first define our variables involved in this analysis (these match the script mentioned above).

```python
motifs_dict = {
    'Zld': 'CAGGTAG',
    'Zld-lowaff': 'CAGGTA',
    'Zld-lowestaff': 'TAGGTAG',
    'Gaf': 'GAGAGAGAGAGAGAGAG',
    'Cad': 'TTTTATGGCC',
    'Dl': 'GGGAAAACCC',
    'Twi': 'AACACATGTT',
    'Bcd': 'TTAATCC'
}
timepoints = ['1to15','15to2','2to25','25to3']
motifs = motifs_dict.keys()
measurement_window_around_motif = 200
primary_position = 300
step = 1
distances_dict = {
             '0to100': list(range(primary_position + 20, primary_position + 100, step)),
             '100to200': list(range(primary_position + 100, primary_position + 200, step)),
             '200to300': list(range(primary_position + 200, primary_position + 300, step)),
             '300to400': list(range(primary_position + 300, primary_position + 400, step))
             }
distance_cat = ChainMap(dict.fromkeys(range(0, 70), 'protein-range'),
                       dict.fromkeys(range(70, 200), 'mononucleosome-range'),
                       dict.fromkeys(range(200, 400), 'enhancer-range'))
distance_cat_list = ['protein-range','mononucleosome-range','enhancer-range']
```

Next, we will average the values, then compute the chromatin-based synergy based on Avsec etal 2021 Nature Genetics. This is slightly different from binding, but because of shoulder-correction terms, we are able to compute single-task (in this case, accessibility is the single task) synergy between multiple motif pairs. 

```python
#%%capture
summary_df = pd.DataFrame()
injection_paths = glob.glob(f'tsv/perturbs/accessibility/insilico/in_silico_injections_paired_profiles_timepoint_*.pkl')
for injection_path in tqdm(injection_paths):
    #Isolate information
    motifA = re.search('motifA_(.*)_motifB', injection_path).group(1)
    motifB = re.search('_motifB_(.*)_distance', injection_path).group(1)
    d_name = re.search('_distance_(.*).pkl', injection_path).group(1)
    t = re.search('timepoint_(.*)_motifA', injection_path).group(1)

    #Load data
    with open(injection_path, 'rb') as f:
        pred_dict = pickle.load(f)

    #Summarize data
    df = summarize_injected_seq(pred_dict,
                                    primary_motif = motifA,
                                    primary_position = primary_position,
                                    secondary_motif = motifB,
                                    secondary_positions = distances_dict[d_name],
                                    measurement_window_around_motif = measurement_window_around_motif)

    #Annotate data
    df['timepoint'] = t
    df['all_synergy_A_corrected'] = (df['all_sum_AB'] - (df['all_sum_B'] - df['all_sum_null'])) / df['all_sum_A']
    df['all_synergy_B_corrected'] = (df['all_sum_AB'] - (df['all_sum_A'] - df['all_sum_null'])) / df['all_sum_B']
    df['windowA_synergy_A_corrected'] = (df['windowA_sum_AB'] - (df['windowA_sum_B'] - df['windowA_sum_null'])) / df['windowA_sum_A']
    df['windowA_synergy_B_corrected'] = (df['windowA_sum_AB'] - (df['windowA_sum_A'] - df['windowA_sum_null'])) / df['windowA_sum_B']
    df['windowB_synergy_A_corrected'] = (df['windowB_sum_AB'] - (df['windowB_sum_B'] - df['windowB_sum_null'])) / df['windowB_sum_A']
    df['windowB_synergy_B_corrected'] = (df['windowB_sum_AB'] - (df['windowB_sum_A'] - df['windowB_sum_null'])) / df['windowB_sum_B']

    df['windowA_maxenh_A_corrected'] = (df['windowA_max_AB'] - (df['windowA_max_B'] - df['windowA_max_null'])) / df['windowA_max_A']
    df['windowA_maxenh_B_corrected'] = (df['windowA_max_AB'] - (df['windowA_max_A'] - df['windowA_max_null'])) / df['windowA_max_B']
    df['windowB_maxenh_A_corrected'] = (df['windowB_max_AB'] - (df['windowB_max_B'] - df['windowB_max_null'])) / df['windowB_max_A']
    df['windowB_maxenh_B_corrected'] = (df['windowB_max_AB'] - (df['windowB_max_A'] - df['windowB_max_null'])) / df['windowB_max_B']

    df['distance'] = df['secondary_position'] - df['primary_position']
    df['distance_cat'] = df['distance'].map(distance_cat.get)

    summary_df = summary_df.append(df)
```

```python
summary_df.to_csv('tsv/perturbs/accessibility/insilico/in_silico_injections_paired_profiles.tsv.gz', sep = '\t', index = False)
```

# Compute Zelda marginalization scores

We want to investigate the role of Zelda accessibility effects in response to varying motif sequences. In order to do this, we will marginalize the motifs by:

1. Measure average log counts of the background accessibility of Zelda over 64 random sequences.
2. Measure average log counts of Zelda-induced accessibility of (1) with a Zelda motif injected.
3. (2) - (1) = marginalization score

First, from `4_*.Rmd`, we grabbed the sequences that were found to be important for Zelda binding according to the BPNet models.

```python
seqs_df = pd.read_csv('tsv/binding_zld_seq_freqs.tsv', sep = '\t')
seqs_df.loc[len(seqs_df.index)] = ['TATCGAT', 0] #Add a negative control based on Nien et al., PLOS Genetics (2011)
seqs_df.head()
```

Next, inject the motifs into the random sequences and measure both values.

```python
%%script false --no-raise-error
marg_df = pd.DataFrame()

for t in timepoints:
    print(t)
    
    #Import model
    tf.keras.backend.clear_session()
    model_path = f'{model_dir}_{t}.h5'
    bias_model_path = f'{model_dir}_{t}.adjusted_bias_model.h5'
    model_chrombpnet, model_bias = load_chrombpnet(model_path = model_path, bias_model_path = bias_model_path)

    for i,row in tqdm(seqs_df.iterrows()):

        #Inject motif into random sequences
        input_flank = (input_seqlen - output_seqlen) // 2
        input_primary_position = 500 + input_flank
        injected_seqs = [insert_motif(seq = s, motif = row['seq'], position = input_primary_position) for s in random_seqs]
        injected_seqs_1he = one_hot_encode_sequences(injected_seqs)

        #Measure predictions over random sequences
        _, inj_wo_motif_preds_arr = predict_chrombpnet(model_chrombpnet, model_bias, random_seqs_1he, no_bias = True, batch_size = 32)
        _, inj_w_motif_preds_arr = predict_chrombpnet(model_chrombpnet, model_bias, injected_seqs_1he, no_bias = True, batch_size = 32)
        
        row['random_log_counts'] = np.log(np.sum(inj_wo_motif_preds_arr, axis = (0,1)) + 1)
        row['injected_log_counts'] = np.log(np.sum(inj_w_motif_preds_arr, axis = (0,1)) + 1)

        #Measure marginalization scores
        row['marg_score'] = row['injected_log_counts'] - row['random_log_counts']
        row['timepoint']=t
        
        #Append to overall pd.df
        marg_df = marg_df.append(row)
marg_df.to_csv('tsv/perturbs/accessibility/accessibility_zld_seq_freqs.tsv.gz', sep = '\t', index = False)
```
